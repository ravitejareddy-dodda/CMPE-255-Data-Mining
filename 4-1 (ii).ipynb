```python
from sklearn.preprocessing import scale # Data scaling
from sklearn import decomposition #PCA
import pandas as pd # pandas
```

```python
X = scale(X)
```

```python
var = decomposition.PCA(2) # we need 2 pca.
pca = var.fit_transform(iris.data)
 
pca.shape
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize = (10, 6))
c_map = plt.cm.get_cmap('jet', 10)
plt.scatter(pca[:, 0], pca[:, 1], s = 15,
            cmap = c_map , c = iris.target)
plt.colorbar()
plt.xlabel('PC-1') , plt.ylabel('PC-2')
plt.show()
```

```python
print('n_components = 2 ', np.cumsum(var.explained_variance_ratio_ * 100)[1])
```

```python
res = decomposition.PCA(3)
Data = res.fit_transform(iris.data)
 
Data.shape
```

```python
print('n_component = 3 ', np.cumsum(res.explained_variance_ratio_ * 100)[2])
```

```python
x = px.scatter_3d(
    Data, x=0, y=1, z=2,
    color=iris.target
)
x.update_traces(marker_size=4)
```
